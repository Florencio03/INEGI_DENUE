```python
def read_csv_to_dict(filename, key_column, value_column):
    data_dict = {}
    with open(filename, newline='', encoding='latin-1') as file:
        next(file)
        csv_reader = csv.reader(file)
        for row in csv_reader:
            key = row[key_column]
            value = row[value_column]
            data_dict[key] = value
    return data_dict
```

```python
def read_csv_to_dict(filename, key_column, value_column):
    data_dict = {}
    with open(filename,  newline='', encoding='latin-1') as file:
        next(file)
        csv_reader = csv.reader(file)
        for row in csv_reader:
            key = row[key_column]
            value = row[value_column]
            if key not in data_dict:
                data_dict[key] = value
    return data_dict

# Example usage:
filename = './conjunto_de_datos/denue_inegi_31_.csv'
key_column_index = 4  # Index of the key column (0-based)
value_column_index = 5  # Index of the value column (0-based)
data = read_csv_to_dict(filename, key_column_index, value_column_index)
#print(data)
```

Definí una función para formar el diccionario, para este caso estoy manejando la columna de los id como las llaves, a su vez cada id forma un diccionario. 

```python
def csv_to_dict(csv_file):
    result = {}
    with open(csv_file, newline='', encoding='latin-1') as file:
        reader = csv.DictReader(file)
        for row in reader:

            key = row.pop(reader.fieldnames[0])
            result[key] = row
    return result
```

```python
csv_file = './conjunto_de_datos/denue_inegi_31_.csv'
my_dict = csv_to_dict(csv_file)
#print(my_dict.keys())
print(type(my_dict))
```

```python
# Initialize lists to store keys and values
all_keys = []
all_values = []

# Function to flatten nested dictionary
def flatten_dict(d):
    for key, value in d.items():
        all_keys.append(key)
        if isinstance(value, dict):
            flatten_dict(value)  # Recursive call for inner dictionary
        else:
            all_values.append(value)

# Flatten the nested dictionary
flatten_dict(acts_by_ageb)
```
